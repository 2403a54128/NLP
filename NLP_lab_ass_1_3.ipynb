```python
text = """
Artificial Intelligence is transforming the world. From healthcare to education,
AI systems are improving efficiency and enabling new possibilities.
"""
```

```python
words = text.split()
word_count = len(words)

print("Number of words:", word_count)
```

```python
lowercase_text = text.lower()

print("Lowercase Text:\n")
print(lowercase_text)
```

```python
print("Original Text:\n", text)
print("\nLowercase Text:\n", lowercase_text)
print("\nTotal Word Count:", word_count)
```
